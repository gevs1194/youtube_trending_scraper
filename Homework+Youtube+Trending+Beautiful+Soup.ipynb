```python
import requests as r
from BeautifulSoup import *
import json
```

```python
url = 'https://www.youtube.com/feed/trending'
re=r.get(url)
data=re.text
soup=BeautifulSoup(data)
```

```python
from pprint import pprint

X=soup.findAll('ul', attrs={'class':"expanded-shelf-content-list has-multiple-items"})
trending=[]
for j in range(len(X)):
    h3=X[j].findAll('h3')
    div=X[j].findAll('div', attrs={'class':"yt-lockup-byline "})
    ul=X[j].findAll('ul',attrs={'class':"yt-lockup-meta-info"})
    for i in range(len(h3)):
        info={}
        link=h3[i].find('a').get('href')
        title=h3[i].find('a').get('title')
        duration=str(h3[i].find('span').text)
        user=div[i].find('a').text
        channel=div[i].find('a').get('href')
        views=ul[i].findAll('li')[1].text
        info['link'] = "https://www.youtube.com"+str(link)
        info['video'] = title
        info['duration'] = duration[11:-1] + " " + "minutes"
        info['user'] = user
        info['channel'] = str(channel)[6:]
        info['views']  = views
        trending.append(info)
pprint(trending)
```

```python
with open('youtube_trending.json', 'w') as output:
    json.dump(trending, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv
```

```python
with open('youtube_trending.csv', 'w') as answers_csv:
    for i in trending:
            writer = csv.writer(answers_csv)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
