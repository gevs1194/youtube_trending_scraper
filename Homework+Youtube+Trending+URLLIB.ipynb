```python
import urllib2 as U2
from lxml import html
from lxml.html import document_fromstring
import json
```

```python
url = 'https://www.youtube.com/feed/trending'
response = U2.urlopen(url)
page = response.read()
tree = document_fromstring(page)
```

```python
from pprint import pprint

X=tree.xpath('//ul[@class="expanded-shelf-content-list has-multiple-items"]')
trending=[]
for j in range(len(X)):
    h3=X[j].cssselect('h3')
    yt=X[j].cssselect('[class="yt-lockup-byline "]')
    ytl=X[j].cssselect('[class="yt-lockup-meta-info"]')
    for i in range(len(h3)):
        info={}
        link=h3[i].cssselect('a')[0].get('href')
        title=h3[i].cssselect('a')[0].text_content()
        duration=str(h3[i].cssselect('span')[0].text_content())
        user=yt[i].cssselect('a')[0].text_content()
        channel=yt[i].cssselect('a')[0].get("href")
        views=ytl[i].cssselect('li')[1].text_content()
        info['link'] = "https://www.youtube.com"+str(link)
        info['video'] = title
        info['duration'] = duration[12:-1] + " " + "minutes"
        info['user'] = user
        info['channel'] = str(channel)[6:]
        info['views']  = views[:-6]
        trending.append(info)
pprint(trending)
```

```python
with open('youtube_trending_xml.json', 'w') as output:
    json.dump(trending, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv
```

```python
with open('youtube_trending_xml.csv', 'w') as csv_answers_1:
    for i in trending:
            writer = csv.writer(csv_answers_1)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
